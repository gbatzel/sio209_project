# Time Series Project - Regression models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# set up pandas and seaborn environments
pd.set_option('display.max_rows', 25)
sns.set()
sns.set_context('notebook')
sns.set_palette('colorblind')
```

### Functions

```python
# Format date and add columns to dataframe
def Organize_Data(i):
    i['Date'] = pd.to_datetime(i['Date'])
    i['YEAR'] = [x.year for x in i.Date]
    i['MONTH'] = [x.month for x in i.Date]
```

```python
# Regression analysis using numpy to find best fit
# p gives the coefficients for regression equation
def Regression_Plot(i, q, g):
    x1 = np.linspace(1,12)
    p = np.polyfit(i.MONTH, i.Temp, q)
    y1 = np.polyval(p, x1)
    plt.scatter(i.MONTH, i.Temp, color='%s' % g)
    plt.plot(x1, y1, '-', color='%s' % g)
```

```python
# Regression analysis using seaborn to find best fit
# Does not give you coefficients for regression equation
def Seaborn_Regression_Plot(i, g):
    sns.regplot(x='MONTH', y='Temp', data=i, order=5, color='%s' % g,
                 x_estimator=np.mean);
```

### Import data

```python
# Import data as dataframe
df = pd.read_csv('../data/san_diego.csv')
df2 = pd.read_csv('../data/santa_cruz.csv')
df3 = pd.read_csv('../data/bird_rock.csv')
df4 = pd.read_csv('../data/SD_ledge.csv')
```

```python
# Format dataframes using Organized_Data function
Organize_Data(df)
Organize_Data(df2)
Organize_Data(df3)
Organize_Data(df4)
```

### Regression analysis using numpy and seaborn

#### San Diego

```python
Regression_Plot(df, 5, 'darkred');
```

```python
Seaborn_Regression_Plot(df, 'darkred');
```

#### Santa Cruz

```python
Regression_Plot(df2, 5, 'darkblue');
```

```python
Seaborn_Regression_Plot(df2, 'darkblue');
```

#### Bird Rock

```python
Regression_Plot(df3, 5, 'darkgreen');
```

```python
Seaborn_Regression_Plot(df3, 'darkgreen');
```

#### San Diego (Ledge)

```python
Regression_Plot(df4, 5, 'lightblue');
```
